```python
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# =========================
# LOAD DATA
# =========================
df = pd.read_csv("nifty_final_dataset.csv")

df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Target variable (stationary already)
y = df["target"]

# =========================
# BASIC PLOT
# =========================
plt.figure(figsize=(10,4))
plt.plot(df["date"], y)
plt.title("Target Series (Stationary)")
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()

# =========================
# ACF PLOT
# =========================
plt.figure(figsize=(10,4))
plot_acf(y, lags=40)
plt.title("ACF Plot")
plt.show()

# =========================
# PACF PLOT
# =========================
plt.figure(figsize=(10,4))
plot_pacf(y, lags=40, method="ywm")
plt.title("PACF Plot")
plt.show()
```
